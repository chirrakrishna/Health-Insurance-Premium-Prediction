<span style="font-size:40px; font-family:Times New Roman fount; color:olive;">Final All Algorithms Result</span>

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, roc_auc_score


results = []

# Define threshold for "high premium"
threshold = 15000
y_test_class = (y_test > threshold).astype(int)

# Linear Regression
yp_lr = lr.predict(X_test)
yp_lr_class = (yp_lr > threshold).astype(int)
results.append([
    'Linear Regression',
    mean_absolute_error(y_test, yp_lr),
    mean_squared_error(y_test, yp_lr),
    np.sqrt(mean_squared_error(y_test, yp_lr)),
    r2_score(y_test, yp_lr),
    roc_auc_score(y_test_class, yp_lr_class)
])

# Decision Tree
yp_dt = dt.predict(X_test)
yp_dt_class = (yp_dt > threshold).astype(int)
results.append([
    'Decision Tree',
    mean_absolute_error(y_test, yp_dt),
    mean_squared_error(y_test, yp_dt),
    np.sqrt(mean_squared_error(y_test, yp_dt)),
    r2_score(y_test, yp_dt),
    roc_auc_score(y_test_class, yp_dt_class)
])

# Support Vector Regressor
yp_svr = svm.predict(X_test)
yp_svr_class = (yp_svr > threshold).astype(int)
results.append([
    'Support Vector Regressor',
    mean_absolute_error(y_test, yp_svr),
    mean_squared_error(y_test, yp_svr),
    np.sqrt(mean_squared_error(y_test, yp_svr)),
    r2_score(y_test, yp_svr),
    roc_auc_score(y_test_class, yp_svr_class)
])

# Random Forest
yp_rf = rf.predict(X_test)
yp_rf_class = (yp_rf > threshold).astype(int)
results.append([
    'Random Forest',
    mean_absolute_error(y_test, yp_rf),
    mean_squared_error(y_test, yp_rf),
    np.sqrt(mean_squared_error(y_test, yp_rf)),
    r2_score(y_test, yp_rf),
    roc_auc_score(y_test_class, yp_rf_class)
])

# XGBoost
yp_xgb = gb.predict(X_test)
yp_xgb_class = (yp_xgb > threshold).astype(int)
results.append([
    'XGBoost',
    mean_absolute_error(y_test, yp_xgb),
    mean_squared_error(y_test, yp_xgb),
    np.sqrt(mean_squared_error(y_test, yp_xgb)),
    r2_score(y_test, yp_xgb),
    roc_auc_score(y_test_class, yp_xgb_class)
])

# Gradient Boosting
yp_gb = gb.predict(X_test)
yp_gb_class = (yp_gb > threshold).astype(int)
results.append([
    'Gradient Boosting',
    mean_absolute_error(y_test, yp_gb),
    mean_squared_error(y_test, yp_gb),
    np.sqrt(mean_squared_error(y_test, yp_gb)),
    r2_score(y_test, yp_gb),
    roc_auc_score(y_test_class, yp_gb_class)
])

results_df = pd.DataFrame(results, columns=['Model', 'MAE', 'MSE', 'RMSE', 'RÂ² Score', 'ROC AUC Score'])
results_df[['MAE', 'MSE', 'RMSE', 'RÂ² Score', 'ROC AUC Score']] = results_df[['MAE', 'MSE', 'RMSE', 'RÂ² Score', 'ROC AUC Score']].round(2)
results_df
```
